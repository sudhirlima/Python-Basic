### Question 1: What is Dimensionality Reduction? Why is it important in machine learning?

**Answer:**  
Dimensionality reduction is the process of reducing the number of input variables or features in a dataset while retaining as much important information as possible. It transforms high-dimensional data into a lower-dimensional space so that patterns, trends, and relationships in the data can be understood more easily.

In other words, itâ€™s a technique used to simplify data without losing significant information.

---

#### **Why Dimensionality Reduction is Important:**
1. **Improves Model Performance:**  
   High-dimensional data can make machine learning models complex and slower to train. Reducing the number of features can lead to faster computation and better generalization.

2. **Reduces Overfitting:**  
   When there are too many features, the model might start learning noise rather than useful patterns. Fewer dimensions help reduce overfitting.

3. **Simplifies Visualization:**  
   Human beings can only visualize data up to 3 dimensions easily. Dimensionality reduction allows us to project data into 2D or 3D for visualization and pattern recognition.

4. **Removes Multicollinearity:**  
   It helps eliminate correlated or redundant features, making the dataset cleaner and more efficient.

5. **Improves Storage and Efficiency:**  
   Lower-dimensional data requires less storage space and computational power.

---

#### **Example:**
Suppose you have a dataset with 1000 features (columns). Many of these features might be correlated or irrelevant. Dimensionality reduction methods like PCA (Principal Component Analysis) can transform those 1000 features into just 10 or 20 while keeping most of the original information.

---

#### **Visualization Example using Python (Optional)**


```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

# Load the dataset
data = load_iris()
X = data.data
y = data.target

# Apply PCA to reduce dimensions to 2
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)

# Plot the reduced data
plt.figure(figsize=(8,6))
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y, cmap='viridis', edgecolor='k', s=80)
plt.title('PCA on Iris Dataset (2D Projection)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

---
**In summary:** Dimensionality reduction simplifies data, enhances learning efficiency, and enables better visualization and interpretation without sacrificing much accuracy.

### Question 2: Name and briefly describe three common dimensionality reduction techniques.

**Answer:**  
There are several techniques used in machine learning to reduce dimensions. The three most common dimensionality reduction techniques are:

---

#### **1. Principal Component Analysis (PCA):**
**Definition:** PCA is a statistical method that converts a set of possibly correlated variables into a set of linearly uncorrelated variables called *principal components*.

**How it works:**
- PCA identifies directions (components) that maximize the variance in the data.
- The first few components capture the majority of the information.

**Example:**
If a dataset has 100 features, PCA might find that 95% of the variance is explained by only 10 components.

**Use Case:** Image compression, noise reduction, feature extraction.

---

#### **2. Linear Discriminant Analysis (LDA):**
**Definition:** LDA is a supervised technique that reduces dimensions while preserving as much class-discriminatory information as possible.

**How it works:**
- It projects the data in a way that maximizes the separation between multiple classes.
- Unlike PCA, LDA considers class labels.

**Use Case:** Face recognition, document classification, and bioinformatics.

---

#### **3. t-Distributed Stochastic Neighbor Embedding (t-SNE):**
**Definition:** t-SNE is a nonlinear technique used for visualizing high-dimensional data in 2D or 3D space.

**How it works:**
- It models the similarity between data points as probabilities.
- It minimizes the difference between these probabilities in high-dimensional and low-dimensional spaces.

**Use Case:** Data visualization in high-dimensional datasets (e.g., MNIST handwritten digits, genomics data).

---


```python
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Load dataset
data = load_digits()
X = data.data
y = data.target

# Apply PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Apply t-SNE
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

# Plot PCA and t-SNE
fig, ax = plt.subplots(1, 2, figsize=(14, 6))
ax[0].scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', s=15)
ax[0].set_title('PCA Visualization')
ax[1].scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis', s=15)
ax[1].set_title('t-SNE Visualization')
plt.show()
```

---
**In summary:**  
- **PCA** focuses on variance preservation,  
- **LDA** focuses on class separation,  
- **t-SNE** focuses on visualizing complex, nonlinear relationships in data.

### Question 3: What is clustering in unsupervised learning? Mention three popular clustering algorithms.

**Answer:**  
Clustering is a key concept in **unsupervised learning**, where the goal is to group similar data points into clusters based on their characteristics â€” without using predefined labels. It helps in identifying patterns, relationships, and hidden structures in data.

For example, clustering can group customers with similar buying habits or segment regions with similar weather patterns.

---

#### **How Clustering Works:**
1. The algorithm analyzes the dataset and identifies groups (clusters) where data points within the same cluster are more similar to each other than to those in other clusters.
2. The similarity is usually measured using metrics like **Euclidean distance**, **Manhattan distance**, or **cosine similarity**.

---

### **Three Popular Clustering Algorithms:**

#### **1. K-Means Clustering:**
**Definition:** K-Means divides data into *k* clusters by minimizing the variance within each cluster.

**How it works:**
- Choose the number of clusters `k`.
- Randomly initialize cluster centers (centroids).
- Assign points to the nearest centroid.
- Recalculate centroids and repeat until convergence.

**Use Case:** Market segmentation, image compression, and anomaly detection.

---

#### **2. Hierarchical Clustering:**
**Definition:** Builds a tree-like structure (dendrogram) of clusters by either merging or splitting existing clusters.

**Types:**
- **Agglomerative:** Start with individual points and merge them step by step.
- **Divisive:** Start with one cluster and split it recursively.

**Use Case:** Document grouping, taxonomy generation, and biological data analysis.

---

#### **3. DBSCAN (Density-Based Spatial Clustering of Applications with Noise):**
**Definition:** A density-based algorithm that groups points close to each other and marks points in low-density regions as outliers.

**How it works:**
- Defines clusters based on density (number of points within a radius).
- Can find arbitrarily shaped clusters and detect noise.

**Use Case:** Geographical data analysis, fraud detection, and noise-resistant clustering.

---


```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Generate synthetic data
X, y = make_blobs(n_samples=300, centers=3, random_state=42)

# Apply K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
y_kmeans = kmeans.fit_predict(X)

# Plot clusters
plt.figure(figsize=(8,6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=200, marker='X')
plt.title('K-Means Clustering Visualization')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

---
**In summary:**  
Clustering helps in discovering natural groupings in data.  
- **K-Means** â†’ Efficient and simple, but assumes spherical clusters.  
- **Hierarchical** â†’ Great for visual insights (dendrograms).  
- **DBSCAN** â†’ Detects arbitrary shapes and noise.

### Question 4: Explain the concept of anomaly detection and its significance.

**Answer:**  
Anomaly detection (also called **outlier detection**) is a technique used to identify rare events, observations, or data points that deviate significantly from the majority of the data. These anomalies can indicate critical incidents such as fraud, equipment failure, network intrusion, or data errors.

In **unsupervised learning**, anomaly detection plays an important role in analyzing unlabeled data to find unexpected patterns or outliers that may require further investigation.

---

### **Why Anomaly Detection is Important:**
1. **Fraud Detection:** Detects unusual transaction patterns in banking or e-commerce systems.
2. **Network Security:** Identifies suspicious activities such as hacking or DDoS attacks.
3. **Manufacturing:** Detects equipment malfunctions or defects in production lines.
4. **Healthcare:** Recognizes abnormal medical conditions or unusual patient data.
5. **Data Cleaning:** Helps remove corrupted or noisy data points before analysis.

---

### **How Anomaly Detection Works:**
- **Step 1:** Analyze the distribution or density of the data.
- **Step 2:** Identify data points that are far from the majority (using distance, probability, or reconstruction error).
- **Step 3:** Flag those points as anomalies or outliers.

Some methods include:
- **Statistical methods** (Z-score, IQR)
- **Distance-based methods** (KNN)
- **Machine learning-based methods** (Isolation Forest, One-Class SVM, Autoencoders)

---


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Generate synthetic data
rng = np.random.RandomState(42)
X = 0.3 * rng.randn(200, 2)
X_outliers = rng.uniform(low=-4, high=4, size=(20, 2))
X = np.r_[X + 2, X - 2, X_outliers]

# Fit Isolation Forest model
model = IsolationForest(contamination=0.1, random_state=42)
model.fit(X)
y_pred = model.predict(X)

# Plot the results
plt.figure(figsize=(8,6))
colors = np.array(['#377eb8', '#ff7f00'])  # Blue = normal, Orange = anomaly
plt.scatter(X[:, 0], X[:, 1], s=60, color=colors[(y_pred + 1)//2], edgecolor='k')
plt.title('Anomaly Detection using Isolation Forest')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

---
### **Explanation of the Code:**
- **Synthetic data** is generated with both normal and anomalous points.
- **Isolation Forest** isolates anomalies instead of profiling normal data.
- Points predicted as `-1` are anomalies, while `1` are normal points.
- The scatter plot visually highlights anomalies in orange.

---

### **In Summary:**
- Anomaly detection helps detect **rare or suspicious patterns** in data.
- Itâ€™s essential for safety, fraud prevention, and data reliability.
- Modern algorithms like **Isolation Forest**, **One-Class SVM**, and **Autoencoders** are widely used for robust anomaly detection.

### Question 5: List and briefly describe three types of anomaly detection techniques.

**Answer:**  
Anomaly detection techniques can be broadly categorized into three main types â€” **Statistical**, **Distance-based**, and **Machine Learning-based** approaches. Each uses a different principle to detect outliers in data.

---

### **1. Statistical Methods (e.g., Z-score, IQR)**
**Concept:**  
These techniques assume that normal data points follow a specific statistical distribution (such as Gaussian). Points that deviate significantly from this distribution are considered anomalies.

**Example:** Z-score Method  
- Z-score measures how far a data point is from the mean in terms of standard deviations.  
- A threshold (e.g., |Z| > 3) identifies points as anomalies.


```python
import numpy as np

# Generate data
np.random.seed(42)
data = np.random.randn(100)
data[95:] = data[95:] + 5  # Inject anomalies

# Compute Z-score
z_scores = (data - np.mean(data)) / np.std(data)
anomalies = np.where(np.abs(z_scores) > 3)

print("Anomalies detected at indices:", anomalies[0])
```

---
### **2. Distance-Based Methods (e.g., KNN Outlier Detection)**
**Concept:**  
These methods measure the distance between data points. A point is considered an anomaly if itâ€™s far from its nearest neighbors.

**Example:** KNN-based Outlier Detection  
- Computes distance to k-nearest neighbors.  
- Points with larger average distances are flagged as anomalies.


```python
from sklearn.neighbors import NearestNeighbors
import numpy as np

# Generate data
X = np.random.randn(100, 2)
X[90:] += 5  # Inject anomalies

# Fit KNN model
knn = NearestNeighbors(n_neighbors=5)
knn.fit(X)
distances, _ = knn.kneighbors(X)

# Average distance
avg_dist = distances.mean(axis=1)
thresh = np.percentile(avg_dist, 95)
anomalies = np.where(avg_dist > thresh)

print("Detected anomalies:", anomalies[0])
```

---
### **3. Machine Learning-Based Methods (e.g., Autoencoder)**
**Concept:**  
Machine learning models learn normal patterns in the data. When the model encounters data points it cannot reconstruct or predict well, those are flagged as anomalies.

**Example:** Autoencoder-based Anomaly Detection  
- Trains a neural network to reconstruct input data.  
- Points with large reconstruction errors are considered anomalies.


```python
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np

# Create synthetic data
X = np.random.randn(300, 2)
X[280:] += 4  # Add anomalies
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Build Autoencoder
model = Sequential([
    Dense(4, activation='relu', input_shape=(2,)),
    Dense(2, activation='relu'),
    Dense(4, activation='relu'),
    Dense(2, activation='sigmoid')
])

model.compile(optimizer='adam', loss='mse')
model.fit(X_scaled, X_scaled, epochs=30, batch_size=16, verbose=0)

# Compute reconstruction error
recon = model.predict(X_scaled)
loss = np.mean(np.square(X_scaled - recon), axis=1)
thresh = np.percentile(loss, 95)
anomalies = np.where(loss > thresh)

print("Detected anomalies:", anomalies[0])
```

---
### **Summary Table:**
| Type | Example | Key Idea | Strengths | Limitations |
|------|----------|-----------|------------|--------------|
| Statistical | Z-score | Deviations from mean | Simple, fast | Assumes normal distribution |
| Distance-based | KNN | Large distances from neighbors | Works without distribution assumption | Sensitive to scaling |
| ML-based | Autoencoder | Reconstruction error | Learns complex patterns | Needs large data, training time |

---

### **In Summary:**
- **Statistical methods** are simple but assume specific data distributions.  
- **Distance-based methods** rely on spatial relationships.  
- **ML-based methods** are powerful and adaptable for complex, high-dimensional data.

### Question 6: What is time series analysis? Mention two key components of time series data.

**Answer:**  
**Time series analysis** is a method of analyzing data points collected or recorded at specific time intervals to identify trends, patterns, and seasonal effects over time. It helps in understanding how data evolves and predicting future values based on past observations.

Time series data are sequential, where the order of observations matters. Examples include stock prices, temperature readings, and sales data recorded over days, months, or years.

---

### **Importance of Time Series Analysis:**
- To forecast future trends and demand.
- To detect patterns and anomalies over time.
- To understand the relationship between time-dependent variables.

---

### **Two Key Components of Time Series Data:**

#### **1. Trend:**
A trend represents the long-term direction in the data (upward, downward, or constant) over a period. It reflects the general tendency of data to increase, decrease, or remain stable.

**Example:** The gradual increase in global average temperature over decades.

#### **2. Seasonality:**
Seasonality represents periodic fluctuations or repeating patterns in data that occur at regular intervals (daily, monthly, or yearly).

**Example:** Ice cream sales peaking during summer months every year.

---


```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic time series data
np.random.seed(42)
time = np.arange(100)
trend = 0.1 * time  # upward trend
seasonality = 10 * np.sin(2 * np.pi * time / 12)  # seasonal component
noise = np.random.normal(scale=2, size=100)

data = trend + seasonality + noise

# Plot the time series
plt.figure(figsize=(10,6))
plt.plot(time, data, label='Time Series Data')
plt.plot(time, trend, label='Trend (0.1 * time)', linestyle='--')
plt.title('Trend and Seasonality in Time Series Data')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()
```

---
### **Explanation of the Code:**
- **Trend:** The gradual upward slope (0.1 * time) shows a long-term increase.
- **Seasonality:** The sinusoidal component repeats periodically.
- **Noise:** Added randomness represents real-world fluctuations.

---

### **In Summary:**
- **Time series analysis** examines time-ordered data to uncover trends and periodic patterns.
- The **two key components** are:
  - **Trend** â†’ Long-term movement.
  - **Seasonality** â†’ Short-term periodic fluctuations.
- These insights help make accurate forecasts and understand time-dependent behavior.

### Question 7: Describe the difference between seasonality and cyclic behavior in time series.

**Answer:**  
In time series data, **seasonality** and **cyclic behavior** are both patterns that repeat over time, but they differ in **duration, regularity, and causes**.

---

### **1. Seasonality:**
- **Definition:** Seasonality refers to **repeating and predictable patterns** in data that occur at **fixed intervals**, such as daily, monthly, quarterly, or yearly.
- **Cause:** Typically influenced by **calendar or seasonal factors** like weather, holidays, or business cycles.
- **Duration:** Occurs in **short, regular, and fixed periods**.
- **Example:** Increase in ice cream sales every summer, higher electricity usage in winter.

**Key Point:** Seasonality is **time-bound and consistent**.

---

### **2. Cyclic Behavior:**
- **Definition:** Cyclic behavior represents patterns that occur over **longer, variable intervals** that are **not strictly periodic**.
- **Cause:** Driven by **economic, social, or business cycles**, not by the calendar.
- **Duration:** The length and amplitude of cycles **vary** and are often **unpredictable**.
- **Example:** Economic recession and recovery cycles, housing market fluctuations.

**Key Point:** Cycles are **irregular and depend on external influences**.

---

### **Comparison Table:**
| Feature | Seasonality | Cyclic Behavior |
|----------|--------------|-----------------|
| **Pattern Type** | Regular and periodic | Irregular and long-term |
| **Cause** | Seasonal or calendar-related | Economic or environmental factors |
| **Duration** | Fixed intervals (e.g., monthly, yearly) | Variable intervals |
| **Predictability** | High | Moderate to low |
| **Example** | Summer sales peak | Business boom and recession |

---


```python
import numpy as np
import matplotlib.pyplot as plt

# Generate time values
time = np.arange(100)

# Create seasonal pattern (fixed period)
seasonal = 10 * np.sin(2 * np.pi * time / 12)

# Create cyclic pattern (variable period)
cyclic = 10 * np.sin(2 * np.pi * time / (20 + 5 * np.sin(time / 10)))

# Plot both patterns
plt.figure(figsize=(12,6))
plt.plot(time, seasonal, label='Seasonal Pattern (Fixed Period)', linewidth=2)
plt.plot(time, cyclic, label='Cyclic Pattern (Variable Period)', linewidth=2)
plt.title('Comparison: Seasonality vs Cyclic Behavior')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

---
### **Explanation of the Code:**
- The **seasonal pattern** repeats consistently every 12 time units (e.g., months).
- The **cyclic pattern** has variable frequency and amplitude, representing irregular cycles.

---

### **In Summary:**
| Aspect | Seasonality | Cyclic Behavior |
|---------|--------------|-----------------|
| **Frequency** | Fixed and regular | Variable and irregular |
| **Predictability** | Easy to model | Hard to predict |
| **Example** | Monthly rainfall patterns | Stock market ups and downs |

**In short:**  
- **Seasonality** = Short-term, predictable, time-bound.  
- **Cyclic behavior** = Long-term, irregular, influenced by complex external factors.

### Question 8: Write Python code to perform K-means clustering on a sample dataset. (Include your Python code and output in the code box below.)

**Answer:**  
K-Means clustering is one of the most popular **unsupervised learning algorithms** used to partition a dataset into *k* distinct, non-overlapping clusters. The algorithm minimizes the distance between data points and their respective cluster centroids.

---

### **How K-Means Works:**
1. Select the number of clusters (**k**).
2. Randomly initialize **k centroids**.
3. Assign each data point to the nearest centroid.
4. Recalculate centroids as the mean of assigned points.
5. Repeat steps 3â€“4 until centroids stabilize or a maximum number of iterations is reached.

---


```python
# Import necessary libraries
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Generate synthetic dataset
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=42)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X)

# Get predicted cluster labels
y_kmeans = kmeans.predict(X)

# Plot the clusters
plt.figure(figsize=(8,6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

# Plot the centroids
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.8, marker='X', label='Centroids')

plt.title('K-Means Clustering Visualization')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

---
### **Explanation of the Code:**
- **`make_blobs()`** creates a synthetic dataset with 4 clusters.
- **`KMeans()`** is initialized with 4 clusters.
- **`fit()`** trains the model, and **`predict()`** assigns each point to a cluster.
- **Centroids** (red X markers) represent the mean position of each cluster.

---

### **In Summary:**
- K-Means is simple, efficient, and effective for well-separated clusters.
- It works best with spherical clusters and numeric data.
- Itâ€™s widely used in **market segmentation**, **image compression**, and **anomaly detection**.

### Question 9: What is inheritance in OOP? Provide a simple example in Python.

**Answer:**  
**Inheritance** is one of the core principles of **Object-Oriented Programming (OOP)**. It allows a new class (called the **child** or **derived** class) to acquire the properties and behaviors (attributes and methods) of an existing class (called the **parent** or **base** class).

Inheritance promotes **code reusability**, **scalability**, and helps in creating a hierarchical structure between classes.

---

### **Types of Inheritance:**
1. **Single Inheritance:** A child class inherits from one parent class.
2. **Multiple Inheritance:** A child class inherits from multiple parent classes.
3. **Multilevel Inheritance:** A class inherits from another class, which in turn is inherited by another class.
4. **Hierarchical Inheritance:** Multiple child classes inherit from a single parent class.
5. **Hybrid Inheritance:** A combination of multiple inheritance types.

---


```python
# Base class (Parent)
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return f"{self.name} makes a sound."

# Derived class (Child)
class Dog(Animal):
    def speak(self):
        return f"{self.name} barks."

# Another derived class
class Cat(Animal):
    def speak(self):
        return f"{self.name} meows."

# Creating objects
obj1 = Dog("Buddy")
obj2 = Cat("Misty")

print(obj1.speak())  # Output: Buddy barks.
print(obj2.speak())  # Output: Misty meows.
```

```python
# Demonstrating use of super() in inheritance
class Puppy(Dog):
    def speak(self):
        parent_speak = super().speak()  # Call parent method
        return f"{parent_speak} (in a cute puppy voice)"

puppy = Puppy("Charlie")
print(puppy.speak())  # Output: Charlie barks. (in a cute puppy voice)
```

---
### **Explanation of the Code:**
- The **`Animal`** class is the **base class** with a `speak()` method.
- The **`Dog`** and **`Cat`** classes inherit from `Animal` and **override** the `speak()` method.
- Each subclass customizes the behavior while reusing the base class constructor.

### **In Summary:**
- Inheritance allows **code reuse** and establishes relationships between classes.
- **Child classes** can override or extend parent class functionality.
- It is a cornerstone concept for building maintainable, modular, and extensible OOP systems.

### Question 10: How can time series analysis be used for anomaly detection?

**Answer:**  
**Time series analysis** can be effectively used for **anomaly detection** by identifying data points that deviate significantly from the expected trend, seasonality, or normal behavior over time.

In real-world applications like stock price monitoring, network traffic, or sensor readings, time series anomalies may represent critical events such as system failures, fraud, or unusual activity.

---

### **How Time Series Analysis Helps in Anomaly Detection:**

1. **Trend and Seasonality Modeling:**  
   Time series decomposition splits data into **trend**, **seasonality**, and **residuals**. Anomalies can be detected when residuals (irregular components) deviate significantly from zero.

2. **Statistical Thresholding:**  
   Compute rolling mean and standard deviation; data points lying outside the confidence interval are flagged as anomalies.

3. **Forecasting Models:**  
   Predict future values using ARIMA, Prophet, or LSTM models. Large differences between predicted and actual values indicate anomalies.

4. **Machine Learning Methods:**  
   Algorithms like Isolation Forest, One-Class SVM, or Autoencoders can detect unusual temporal patterns in sequential data.

---


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create synthetic time series data
np.random.seed(42)
time = np.arange(200)
data = 50 + 0.2 * time + 5 * np.sin(2 * np.pi * time / 25) + np.random.normal(0, 1, 200)

# Inject anomalies
anomalies = np.random.choice(time, 8, replace=False)
data[anomalies] += np.random.uniform(8, 15, size=8)

# Create DataFrame
ts = pd.DataFrame({'Time': time, 'Value': data})

# Calculate rolling mean and standard deviation
rolling_mean = ts['Value'].rolling(window=10, center=True).mean()
rolling_std = ts['Value'].rolling(window=10, center=True).std()

# Define anomaly threshold
upper_bound = rolling_mean + 2 * rolling_std
lower_bound = rolling_mean - 2 * rolling_std

# Detect anomalies
anomaly_points = ts[(ts['Value'] > upper_bound) | (ts['Value'] < lower_bound)]

# Plot results
plt.figure(figsize=(12,6))
plt.plot(ts['Time'], ts['Value'], label='Time Series Data', color='blue')
plt.plot(ts['Time'], rolling_mean, label='Rolling Mean', color='orange', linewidth=2)
plt.fill_between(ts['Time'], lower_bound, upper_bound, color='gray', alpha=0.2, label='Normal Range')
plt.scatter(anomaly_points['Time'], anomaly_points['Value'], color='red', label='Anomalies', s=80)
plt.title('Anomaly Detection using Time Series Analysis')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()
```

---
### **Explanation of the Code:**
- **Synthetic time series** generated with a trend and seasonal component.
- **Rolling mean and standard deviation** used to create an expected range.
- Points outside Â±2 standard deviations are flagged as anomalies.
- Red dots on the plot indicate detected anomalies.

---

### **In Summary:**
| Approach | Description | Example Technique |
|-----------|--------------|-------------------|
| **Statistical** | Identify deviations using z-scores or rolling stats | Rolling mean, Bollinger bands |
| **Forecast-based** | Compare predicted vs. actual values | ARIMA, Prophet |
| **Machine Learning** | Learn normal patterns and flag irregular behavior | LSTM, Isolation Forest |

**In short:** Time series analysis allows detection of **contextual anomalies** by comparing observed values against expected temporal patterns â€” making it a powerful tool for monitoring and predictive maintenance.
